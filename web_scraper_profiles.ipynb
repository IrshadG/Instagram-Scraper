# Scraping top 5k most followed Instagram profile names

Website used: https://starngage.com/


```python
import re
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
```

```python

row_collector = []
progress = tqdm(range(1,51))
for page in progress:
  # download the .html files from the website into the folder 'ProfileIds'
  with open("./ProfileIds/page"+ str(page)+".html",encoding="utf8") as fp:
    soup = BeautifulSoup(fp, 'html.parser')
  tsoup = soup
  rows = tsoup.findAll("tr")
  try:
      for row in range(1,101):
        varName = rows[row].find('div', class_='mb-1 fw-bold').contents
        Name = varName[0].strip() if len(varName)>0 else "NA" 
        profile = rows[row].find('a', class_='link color-pink text-break').contents[0]
#         followers =  rows[row].findAll('div', string=re.compile("M"))[-1].get_text().strip()
        followers =  rows[row].findAll('div', string=re.compile(r'\d+M'))[-1].get_text().strip()
        if rows[row].find('div', class_="country d-flex align-items-center justify-content-center"):
            country = rows[row].find('div', class_="country d-flex align-items-center justify-content-center").a.contents[0].strip() 
        else:
            country = "NA"
        ToI = ', '.join([var.contents[0].strip() for var in rows[row].find('div', class_="d-flex flex-wrap topic-wrap justify-content-center").find_all('a') if len(var.contents)>0])
        temp = [Name, profile,followers, country, ToI]
        row_collector.append(temp)
  except:
      print(row)
        
   # Enter rate limiting per page here
df = pd.DataFrame(row_collector)
df[['Name','Profile','Followers','Country', 'Topic of Influence']] = df[[0,1,2,3,4]]
del df[0]
del df[1]
del df[2]
del df[3]
del df[4]
display(df)
```

```python
df.to_csv('./5kprofile.csv') 
```
